```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
cleaned_test = []
cleaned_stop_test = []
for review in takeaways:
    #print(review, type(review))
    #test2 = " ".join(review)
    print(review)
    review2 = [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in nltk.word_tokenize(review)]
    #print(review2)
    review3 = [i.strip("!\"“'’#$%&()*+-./:;<=>?@[\]^_`{|}~\n012345789") for i in review2]
    #print(review3)
    review4 = [word for word in review3 if len(word) > 1]
    print(review4)
    cleaned_test.append(review4)
```

```python
#TF-IDF; term ferquency * inverse document frequency

#TF(t,d) = (count of terms in document) / (total number of words in documents)
#IDF(t) = N/df; (total number of documents) / (count of occurences of term in document set N)

# for large corpus ---> IDF can become very large, when this happens good to use the log and adding 1 to smooth ---> log(N/(df+1))

```

```python
final_flat_list = [item for sublist in final_reviews2 for item in sublist]
```

```python
term_count = {}
for i in final_flat_list:
    if i not in term_count:
        term_count[i] = 1
    else:
        term_count[i] += 1
        
    
```

```python
final_reviews2
tf_values = []
review_index = 0
for review in final_reviews2:
    temp_counter  = {}
    for word in review:
        if word not in temp_counter:
            temp_counter[word] = 1
        else:
            temp_counter[word] += 1
    for uniques in temp_counter.keys():
        tf_values.append((uniques,temp_counter[uniques], temp_counter[uniques]/len(review), review_index))
    review_index += 1
```

```python
# (word,number of times word appears in review, tf_value, review[index])
tf_values
```

```python
idf_values = []
for i in tf_values:
    idf_values.append((i[0], term_count[i[0]], sum(str(i[0]) in s for s in final_reviews2), np.log(term_count[i[0]]/sum(str(i[0]) in s for s in final_reviews2))))
    
```
